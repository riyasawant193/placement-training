```python
class Stack_Example:
    def __init__(self,size):
        self.tos=-1
        self.slist=[]
        for i in range(size):
            self.slist.append(0)
        self.maxsize=size
    def push(self,element):
        self.tos+=1
        self.slist[self.tos]=element
        print("pushed:",element)
    def pop(self):
        temp=self.slist[self.tos]
        self.tos-=1
        return(temp)
    def isfull(self):
        if self.tos==self.maxsize-1:
            return True
        else:
            return False
    def isempty(self):
        if self.tos==-1:
            return True
        else:
            return False
    def peek(self):
        return(self.slist[self.tos])
    def print_stack(self):
        for i in range(self.tos,-1,-1):
            print(self.slist[i])
```

```python
#menudriven code for stack
'''start with size of stack----user
1.push 2.pop 3.peek 4.print 0.Exit
before push check full or not
before pop check empty of not
before peek check empty of not
before print check empty of not
also check for wrong input
'''

size=int(input("Enter size:"))
obj=Stack_Example(size)
while True:
    ch=int(input("\n1.Push\n2.Pop\n3.Peek\n4.Print\n0.Exit\n:"))
    if ch==1:
        if obj.isfull()!=True:
            e=int(input("Enter element to be pushed:"))
            obj.push(e)
        else:
            print("Stack is full")
    elif ch==2:
        if obj.isempty()!=True:
            e=obj.pop()
            print(e,"is popped")
        else:
            print("Stack is empty")
    elif ch==3:
        if obj.isempty()!=True:
            e=obj.peek()
            print(e,"is at peek")
        else:
            print("Stack is empty")
    elif ch==4:
        if obj.isempty()!=True:
            print("Stack contains the following:")
            obj.print_stack()
        else:
            print("Stack is empty")
    elif ch==0:
         print("Thankyou for using the code")
         break
    else:
        print("Invalid option")
```

```python
class Queue_Example:
    def __init__(self,size):
        self.front,self.rear=0,-1
        self.qlist=[]
        for i in range(size):
            self.qlist.append(0)
        self.maxsize=size
    def enqueue(self,element):
        self.rear+=1
        self.qlist[self.rear]=element
        print("Enqueued:",element)
    def dequeue(self):
        temp=self.qlist[self.front]
        self.front+=1
        return(temp)
    def isfull(self):
        if self.rear==self.maxsize-1:
            return True
        else:
            return False
    def isempty(self):
        if self.rear<self.front:
            return True
        else:
            return False

    def print_queue(self):
        for i in range(self.front,self.rear+1):
            print(self.qlist[i],end="-")
```

```python
size=int(input("Enter size:"))
obj=Queue_Example(size)
while True:
    ch=int(input("\n1.Enqueue\n2.Dequeue\n3.Print\n0.Exit\n:"))
    if ch==1:
        if obj.isfull()!=True:
            e=int(input("Enter element to be added:"))
            obj.enqueue(e)
        else:
            print("Queue is full")
    elif ch==2:
        if obj.isempty()!=True:
            e=obj.dequeue()
            print(e,"is removed")
        else:
            print("Queue is empty")
    elif ch==3:
        if obj.isempty()!=True:
            print("Queue contains the following:")
            obj.print_queue()
        else:
            print("Queue is empty")
    elif ch==0:
         print("Thankyou for using the code")
         break
    else:
        print("Invalid option")
```

```python
class Node:
	def __init__(self,data):
		self.data=data
		self.next=None#no data type

class Linkedlist:
    def createLinkedlist(self):
          self.root=None
        #Root is not created but assigned
    def insertLeft(self,e):
        n=Node(e)
        if self.root==None:
            self.root=n#if not there assign n as root
        else:
            n.next=self.root#1
            self.root=n#2
        print(e,"added to list")

    def deleteLeft(self):
        if self.root==None:
            print("Empty List")
        else:
            t=self.root#1
            self.root=self.root.next#2
            print(t.data,"deleted from list")

    def insertRight(self,e):
        n=Node(e)
        if self.root==None:
            self.root=n#if not there assign n as root
        else:
            t=self.root#1
            while t.next!=None:#2
                t=t.next
            t.next=n#3
        print(e,"added to list")

    def deleteRight(self):
        if self.root==None:
            print("List Empty")
        else:
            t2=t=self.root#1
            while t.next!=None:#2
                t2=t
                t=t.next
            if self.root==t:
                self.root=None
            else:
                t2.next=None#3
            print(t.data,"deleted from list")

    def printList(self):
        if self.root==None:
            print("List Empty")
        else:
            t=self.root#1
            while t!=None:#2
                print(t.data,end='->')
                t=t.next

    def searchList(self,key):
        if self.root==None:
            print("List Empty")
        else:
            t=self.root#1
            while t!=None:#2
                if key==t.data:
                    print(key,"Element found")
                    break
                t=t.next
            else:#only executes if while runs completely
                print("Element not found")


obj=Linkedlist()
obj.createLinkedlist()
while True:
    ch=int(input("\n1.Insert Left\n2.Delete Left\n3.Insert Right\n4.Delete Right\n5.Display\n6.Search\n0.Exit\n:"))
    if ch==1:
        e=int(input("Enter element:"))
        obj.insertLeft(e)
    elif ch==2:
        obj.deleteLeft()
    elif ch==3:
        e=int(input("Enter element:"))
        obj.insertRight(e)
    elif ch==4:
        obj.deleteRight()
    elif ch==5:
        obj.printList()
    elif ch==6:
        e=int(input("Enter element:"))
        obj.searchList(e)
    elif ch==0:
         print("Thankyou for using the code")
         break
    else:
        print("Invalid option")
```
