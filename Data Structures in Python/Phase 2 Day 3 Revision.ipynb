```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

def insert(r,data):
    if r==None:
        return Node(data)
    else:
        if data<r.data:
            r.left=insert(r.left,data)
        elif data>=r.data:
            r.right=insert(r.right,data)
    return r

def inorder(r):#lpr
    if r:
        inorder(r.left)
        print(r.data)
        inorder(r.right)
def preorder(r):#lpr
    if r:
        print(r.data)
        preorder(r.left)
        preorder(r.right)
def postorder(r):#lpr
    if r:
        postorder(r.left)
        postorder(r.right)
        print(r.data)
```

```python
#graph
from collections import deque
class Graph:
    def __init__(self, num_vertices):
        self.num_vertices = num_vertices
        self.adj_matrix = [[0] * num_vertices for _ in range(num_vertices)]
        for source in range(self.num_vertices):
            for dest in range(self.num_vertices):
                self.adj_matrix[source][dest] = int(input("Enter edge value from V"+str(source)+" to V"+str(dest)+" :"))

    def display(self):
        for row in self.adj_matrix:
            print(row)

    def dfs(self, start_vertex):
        visited = [False] * self.num_vertices
        self._dfs_recursive(start_vertex, visited)

    def _dfs_recursive(self, vertex, visited):
        visited[vertex] = True
        print(vertex, end=' ')

        for neighbor in range(self.num_vertices):
            if self.adj_matrix[vertex][neighbor] == 1 and not visited[neighbor]:
                self._dfs_recursive(neighbor, visited)

    def bfs(self, start_vertex):
        visited = [False] * self.num_vertices
        queue = deque()
        queue.append(start_vertex)#enqueue
        visited[start_vertex] = True#visited marked

        while queue:#till data in queue exists
            vertex = queue.popleft()#removed 1st element
            print(vertex, end=' ')

            for neighbor in range(self.num_vertices):
                if self.adj_matrix[vertex][neighbor] == 1 and not visited[neighbor]:
                    queue.append(neighbor)
                    visited[neighbor] = True


# Example usage:
graph = Graph(4)
graph.display()

'''print("DFS:")
graph.dfs(0)
print("\nBFS:")
graph.bfs(0)
'''
```

```python
#create list of n numbers and then call bubble sort to sort the number list
def bubble(a):
	for i in range(len(a)):#passes
		for j in range(len(a)-1):
			if a[ j ]>a[j+1]:
				a[j],a[j+1]=a[j+1],a[j]

dlist=[90,30,50,60,40,10,70,100,20]
print("Following is the list before bubble sort: ",dlist)
bubble(dlist)
print("Following is the list after bubble sort: ",dlist)
```

```python
def selection(a):
    print("Selction activated")
    for i in range(len(a)):
        min=a[i]
        pos=i
        for j in range(i+1,len(a)):
            if a[ j ]<min:
                min=a[ j ]
                pos=j
        a[i],a[pos]=a[pos],a[i]

dlist=[22,11,55,44,33,77,66,99,88]
print("Following is the list before selection sort: ",dlist)
selection(dlist)
print("Following is the list before selection sort: ",dlist)
```

```python
def insertion_sort(a):
    print("Insertion activated")
    for i in range(len(a)-1):
        new_element=a[ i+1]
        j=i+1
        while j>0 and a[ j-1]>new_element:
            a[ j ]=a[ j-1]
            j-=1
            a[ j ]=new_element

dlist=[22,11,55,44,33,77,66,99,88]
print("Following is the list before insertion sort: ",dlist)
insertion_sort(dlist)
print("Following is the list before insertion sort: ",dlist)
```

```python
def quick_sort(a,start,end):
  i=start
  j=end
  pivot=start
  while i<j:
    while a[i]<a[pivot]:
      i+=1
    while a[j]>a[pivot]:
      j-=1
    if i<j:
      a[i],a[j]=a[j],a[i]
    if i<end:
      quick_sort(a,i+1,end)
    if j>start:
      quick_sort(a,start,j-1)

dlist=[22,11,55,44,33,77,66,99,88]
print("Following is the list before quick sort: ",dlist)
quick_sort(dlist,0,len(dlist)-1)
print("Following is the list after quick sort: ",dlist)
```

```python
def merger(a,start,mid,end):
  i=start
  j=mid+1
  temp=[0]*len(a)
  ti=start
  while i<=mid and j<=end:
    if a[i]<a[j]:
      temp[ti]=a[i]
      i+=1
    else:
      temp[ti]=a[j]
      j+=1
    ti+=1
  while i<=mid:
    temp[ti]=a[i]
    i+=1
    ti+=1
  while j<=end:
    temp[ti]=a[j]
    j+=1
    ti+=1
  for i in range(start,end+1):
    a[i]=temp[i]

def merge_sort(a,start,end):
  if start<end:
    mid=(start+end)//2
    merge_sort(a,start,mid)
    merge_sort(a,mid+1,end)
    merger(a,start,mid,end)

dlist=[28,27,6,2,3]
print("Following is the list before quick sort: ",dlist)
merge_sort(dlist,0,len(dlist)-1)
print("Following is the list after quick sort: ",dlist)
```

```python
def sequential_search(alist,key):
  for i in range(len(alist)):
    if alist[i]==key:
      return i
  return -1

dlist=[88,11,33,22,99,77,55,66,44]
key=int(input("Enter key to search: "))
res=sequential_search(dlist,key)
if res==-1:
  print(key,"not found")
else:
  print(key,"found at",res,"position")
```

```python
def binary_search(alist,start,end,key):
  if start<end:
    mid=(start+end)//2
    if key==alist[mid]:
      return mid
    else:
      if key<alist[mid]:
        return binary_search(alist,start,mid-1,key)
      else:
        return binary_search(alist,mid+1,end,key)
  else:
    return (-1)

dlist=[11,22,33,44,55,66,77,88,99]
key=int(input("Enter key to search: "))
res=binary_search(dlist,0,len(dlist)-1,key)
if res==-1:
  print(key,"not found")
else:
  print(key,"found at",res,"position")
```
