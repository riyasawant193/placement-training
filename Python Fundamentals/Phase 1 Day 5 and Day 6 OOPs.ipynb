```python
#calling from the base
class Human:
    def __init__(self,name,gender):
      self.name=name
      self.gender=gender
class Student(Human):
    def __init__(self,name,gender,degree):
      super().__init__(name,gender)
      self.degree=degree
class Job(Student):
    def __init__(self,name,gender,degree,salary):
      super().__init__(name,gender,degree)
      self.salary=salary
    def display(self):
      print(self.name,self.gender,self.degree,self.salary)
```

```python
'''compare car and bike and print who is faster
we have class comparator that has method comapre()
that takes any two data one from car class type and
other from bike class type both inherits class automobile
that gives them property speed
car has extra propery of 4-c engine
bike has extra property of 2-c engine'''

class Automobile:
    def __init__(self,speed):
        self.speed=speed
        print("Speed is set at ",self.speed)
class Car(Automobile):
    def __init__(self,speed,four_c_engine):
      super().__init__(speed)
      self.four_c_engine=four_c_engine
class Bike(Automobile):
    def __init__(self,speed,two_c_engine):
      super().__init__(speed)
      self.two_c_engine=two_c_engine
class Comparator:
    def compare(self,car,bike):
      if car.speed>bike.speed:
        print("Car is faster with ",car.speed)
      else:
        print("Bike is faster with ",bike.speed)
```

```python
c=Car(220,1.6)
b=Bike(380,550)
comp=Comparator()
comp.compare(c,b)
```

```python
#we have object of student class consisting name,gender,pointer
#we have all data in list(to be created),sort this list in
#descending order of pointer

class Student:
    def __init__(self,name,gender,pointer):
        self.name=name
        self.gender=gender
        self.pointer=pointer
    def __str__(self):
        return self.name+"\t"+self.gender+"\t"+str(self.pointer)

students_list=[]

s1=Student("Riya","female",8.9)
students_list.append(s1)
s1=Student("Khushi","female",9.9)
students_list.append(s1)
s1=Student("Parth","male",8.5)
students_list.append(s1)
s1=Student("James","male",8.0)
students_list.append(s1)
s1=Student("Jackson","male",8.1)
students_list.append(s1)

#bubble sort
for i in range(len(students_list)):
      for j in range(len(students_list)-1):
            if students_list[j].pointer<students_list[j+1].pointer:
              students_list[j],students_list[j+1]=students_list[j+1],students_list[j]

for i in students_list:
  print(i)
```

```python
#accept number from user if -ve raise exception NegativeNumber

class NegativeNumber(Exception):
  def __init__(self,number):
    self.number=number
  def __str__(self):
    return "Exception "+str(self.number)+" Please enter a positive number"

try:
  n=int(input("Enter number: "))
  if n>0:
    print("Good job! The number you entered is ",n)
  else:
    raise NegativeNumber(n)
except Exception as msg:
  print(msg)
```
