```python
#Roots of Quadratic Equation
#1. print roots 2. conditions 3. input quadratic equation
import math
a=float(input("Enter the value of a: "))
b=float(input("Enter the value of b: "))
c=float(input("Enter the value of c: "))
root1=(-b-math.sqrt((b**2)-(4*a*c)))/(2*a)
root2=(-b+math.sqrt((b**2)-(4*a*c)))/(2*a)
print("Roots of the quadratic equation are ",root1,root2)
```

```python
#Leap year or not
#Year perfectly div by 4 but not by 100 or can be by 400
#1. print leap year or not 2. condition 3. input year
year=int(input("Enter year:"))
if year%4==0 and year%100!=0 or year%400==0:
    print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```

```python
#-print start to end
    #input: start 2 end is 5 2,3,4,5
    #input: start 5 end is 2 5,4,3,2
start=int(input("Start"))
end=int(input("End"))
if start<end:
    for i in range(start,end+1,+1):
        print(i)
else:
    for i in range(start,end-1,-1):
        print(i)
```

```python
#start to end  but add every odd number and multiply every even number
start=int(input("Start :"))
end=int(input("End :"))
product=1
sum=0
for i in range(start,end+1,1):
    if i%2==0:
        product*=i
    else:
        sum+=i
print("Total is:",sum+product)
```
