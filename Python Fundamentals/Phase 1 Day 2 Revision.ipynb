```python
#accept all elements in list till 1st blank given
#and then find min,max,sum and average of all numbers from list
dataList=[]                   # blank list
while True:
    data=input("Enter data :")
    if data=='':              # stop when blank
        break
    dataList.append(float(data))


print("Data in list:",dataList)
print("Sum of the values in the list is ",sum(dataList))
print("Number of the values in the list is ",len(dataList))
print("Minimum number from the list is ",min(dataList))
print("Maximum number from the list is ",max(dataList))
print("Average is ",sum(dataList)/len(dataList))
```

```python
#accept all elements in list till 1st blank given
#-5 from all number greater than average
#+10 to all numbers lesser than average
datalist=[]                   #blank list
while True:
    data=input("Enter data :")
    if data=='':
        break
    datalist.append(float(data))
print("List after adding 10 from numbers lesser than average",datalist)
average=sum(dataList)/len(dataList)
for i in range(len(datalist)):
  if datalist[i]<average:
    datalist[i]+=10
  else:
    datalist[i]-=5
print("List after subtracting 5 from numbers greater than average",datalist)
```

```python
'''&:intersection
|:union
^:symmetric difference
-:difference'''

s={1,2,3,4,5,6,7,8,9}
sing={2,5,7}
dance={2,4,8}
security={1,2,9}

#one who has not taken part in anything
nonparticipant=s-(sing|dance|security)
print("One who has not taken part in anything is ",nonparticipant)

#one who is part in all
enthusiastic=sing&dance&security
print("One who is part in all is ",enthusiastic)

#one who are part in dance and singing
dance_and_sing=(sing&dance)
print("One who is part in dance and singing is",dance_and_sing)

#one who can be in security who are not in anything
couldbe_security=s-dance-sing
print("One who can be in security who is not in anything is ",couldbe_security)
```

```python
#you are a tea stall owner
#selling tea(10),coffee(20),water(15),bun maska(20)
#allow user to select the order from menu driven
#0.exit
#on exit list amount earned, and list number of tea,coffee and other things sold

ct=cc=cw=cb=0
while True:
  ch=int(input("1. Tea\n2. Coffee\n3. Water\n4. Bun Maska\n0. Exit\n"))
  if ch==1:
        print("Tea Selected")
        ct+=1
  elif ch==2:
        print("Coffee Selected")
        cc+=1
  elif ch==3:
        print("Water Selected")
        cw+=1
  elif ch==4:
        print("Bun Maska Selected")
        cb+=1
  elif ch==0:
        print("Thankyou for ordering ")
        break
  else:
    print("Invalid choice")

total=(ct*10)+(cc*20)+(cw*15)+(cb*20)
print("Total units sold:",(ct+cc+cw+cb))
print("Earned:",total)
```
