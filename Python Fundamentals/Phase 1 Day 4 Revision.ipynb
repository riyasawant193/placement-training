```python
#using function add to list to create averge 3rd list
'''
l1=[10,20,30]
l2=[5,15,25]
l3=[(10+5)/2,(20+15)/2,(30+25)]'''

def listadder(l1,l2):
  temp=[]
  for i in range(len(l1)):
    temp.append((l1[i]+l2[i])/2)
  return temp

l1=[10,20,30]
l2=[5,15,25]
print("The average of the two lists is ",listadder(l1,l2))
```

```python
#find leader in a given list
#leader is the one who has all prior smaller
#expected output: 3,7,9,10 should be in a list produced by function called LeaderGenerator()

#output: list--->condition--->number
#process: ith is a leader iff all other from 0 to i-1 are lesser
#input: list to check on


def LeaderGenerator(list):
    leader=[]
    for i in range(len(list)):
        flag=True
        for j in range(i):
            if list[j]>=list[i]:
                flag=False
            if flag:
                leader.append(list[i])
    return(leader)


list=[3,1,7,2,9,10]
leader_list=LeaderGenerator(list)
print("The leader in the given list is ",leader_list)
```

```python
'''Find Common Elements in Two Lists:
○ Question: Write a function to find common elements between two lists.
○ Logic: Use list comprehensions or a loop to compare elements.
○ Sample Input: [1, 2, 3, 4] and [3, 4, 5, 6]
○ Expected Output: [3, 4]
output: list of common elements in both
process: element to element comparission(in/not in)------use set
         def intersection(l1,l2)
input: 2 lists
'''
def intersection(l1,l2):
    common_list=[]
    for i in l1:
        if i in l2:
            common_list.append(i)
    return common_list
def intersection2(l1,l2):
    return list(set(l1)&set(l2))

l1=[1, 2, 3, 4]
l2=[3, 4, 5, 6]
common=intersection2(l1,l2)
print("common to both:",common)

```

```python
'''
Remove All Occurrences of an Element in a List:
Question: Write a function to remove all occurrences of a specific
element from a list.
Sample

Expected Output: List: [1, 3, 4]
process:count total time it is there,remove each time
                2--count 3---- call remove 3 times
Input: List: [1, 2, 2, 3, 4, 2] and Element: 2
'''

def no_occurences(input_list,n):
  c=input_list.count(n)
  for i in range(c):
    input_list.remove(n)

input_list=[1, 2, 2, 3, 4, 2]
n=2
no_occurences(input_list,n)
print("List without element",n,"is",input_list)
```

```python
#list carry binary number in individual element manner
#convert it to dec
'''input:[1,0,1,0]  output:10'''

def BinarytoDecimal(binary_list):
    decimal=0
    binary_list.reverse()
    for i in range(len(binary_list)):
      decimal+=binary_list[i]*(2**i)
    return(decimal)

binary_list=[1,0,1,0]
decimal_result=BinarytoDecimal(binary_list)
print("The decimal form of the given digit is ",decimal_result)
```

```python
def NonReapeater(list):
  for i in list:
    if list.count(i)==1:
      return(i)

list=[1,2,4,2,1,3,5]
element=NonReapeater(list)
print("First non repeating element is ",element)
```

```python
'''Write a program to check if a given string is a pangram
(contains all letters of the alphabet).
○ Logic: Use loops to check if all letters are present.
○ Sample Input: "The quick brown fox jumps over the lazy dog."'''
#flag=set
#loops from a to z
    #check for each alph if not found reset flag and stop
#check flag and comment

pangram="The quick brown fox jumps over the lazy dog"
flag=True

for i in "abcdefghijklmnopqrstuvwxyz":
    if i not in pangram:
      flag=False
      break
if flag==True:
      print("The given sentence is a pangram")
else:
      print("The given sentence is not a pangram")
```
