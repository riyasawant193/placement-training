```python
#anagram or not listen=silent

fword=input("Enter first word: ")
sword=input("Enter second word: ")
flist=sorted(list(fword))
slist=sorted(list(sword))

if flist==slist:
  print("The given word is an anagram")
else:
  print("The given word is not an anagram")
```

```python
# anagram or not
w1="on"
w2="no"
#characterize this
#sort the list
list_w1=sorted(list(w1))
list_w2=sorted(list(w2))
#compare both character by character
flag=True
for i in range(len(list_w1)):
    if list_w1[i]!=list_w2[i]:
        flag=False
        break
if flag==True:
    print("yes")
else:
    print("no")
#print the result
```

```python
#in place palindrome
s=input("Enter a word :")
flag=True
i,j=0,len(s)-1
while i<=j:
  if s[i]!=s[j]:
    flag=False
    break
  i+=1
  j-=1
if flag==True:
  print("The given word is a palindrome")
else:
  print("The given word is not a palindrome")
```

```python
def min(no1,no2):#returning is better so one can nest it
    if no1<no2:
        return no1
    else:
        return no2
```

```python
def min(no1,no2,no3,no4):
  if (no1<no2 and no1<no3 and no1<no4):
    return no1
  elif (no2<no1 and no2<no3 and no2<no4):
    return no2
  elif (no3<no1 and no3<no2 and no3<no4):
    return no3
  else:
    return no4
```

```python
#using function add to list to create averge 3rd list
'''
l1=[10,20,30]
l2=[5,15,25]
l3=[(10+5)/2,(20+15)/2,(30+25)]'''

def listadder(l1,l2):
  temp=[]
  for i in range(len(l1)):
    temp.append((l1[i]+l2[i])/2)
  return temp

l1=[10,20,30]
l2=[5,15,25]
print("The average of the two lists is ",listadder(l1,l2))
```
